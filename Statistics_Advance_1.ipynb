17. Estimation & Confidence Intervals
Point Estimate: A single best guess for a population parameter.

Interval Estimate (Confidence Interval):

Provides a range where the true parameter likely lies.

Linked to Z-scores.

18. A confidence interval (CI) is a range of values used to estimate an unknown population parameter (e.g., mean or proportion) with a certain level of confidence.

- Incorporate Variability & Uncertainty

Confidence intervals account for sampling variability, showing that repeated samples would give slightly different estimates.

- Help in Decision Making

If a confidence interval for a drug‚Äôs effectiveness does not include zero, researchers can conclude it is likely effective.

If two group means' CIs overlap significantly, there may be no significant difference between them.

A 95% confidence interval corresponds to a 5% significance level in hypothesis testing.

If a null hypothesis value (e.g., a mean difference of 0) is outside the confidence interval, we reject the null hypothesis.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import randint

# Generate discrete uniform distribution (1 to 6)
low, high = 1, 6
samples = randint.rvs(low, high+1, size=10000)

# PMF Calculation
values, counts = np.unique(samples, return_counts=True)
pmf = counts / counts.sum()

# Plot PMF
plt.bar(values, pmf, color='blue', alpha=0.7, label="PMF")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.title("Discrete Uniform Distribution (1-6)")
plt.legend()
plt.show()
```

```python
from scipy.stats import bernoulli

def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)

# Example
p = 0.5
x_values = [0, 1]
pdf_values = [bernoulli_pdf(p, x) for x in x_values]

print("Bernoulli PDF values:", pdf_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

low, high = 1, 6
samples = randint.rvs(low, high+1, size=10000)

# Compute CDF
values, counts = np.unique(samples, return_counts=True)
cdf = np.cumsum(counts) / counts.sum()

# Plot CDF
plt.step(values, cdf, where="post", color="red", label="CDF")
plt.xlabel("Values")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (Discrete Uniform)")
plt.legend()
plt.show()
```

```python
from scipy.stats import norm

# Sample data
sample_mean = 52
pop_mean = 50
pop_std = 10
n = 30

# Compute Z-statistic
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 1 - norm.cdf(z)

print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis.")
```

```python
import numpy as np
import scipy.stats as stats

data = np.random.normal(50, 10, 100)
mean, std = np.mean(data), np.std(data, ddof=1)
confidence = 0.95

margin_error = stats.norm.ppf((1 + confidence) / 2) * (std / np.sqrt(len(data)))
lower, upper = mean - margin_error, mean + margin_error

print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
import numpy as np
import scipy.stats as stats

sample = np.random.normal(100, 15, 50)
mean, std = np.mean(sample), np.std(sample, ddof=1)
conf_int = stats.norm.interval(0.95, loc=mean, scale=std/np.sqrt(len(sample)))

print(f"95% Confidence Interval for Mean: {conf_int}")
```
